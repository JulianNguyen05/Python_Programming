## Thừa kế (Inheritance)


```python
# Định nghĩa lớp cơ sở (Base / Parent / Super class)
class Person:
    def __init__(self, name, age): 
        self.name = name
        self.age = age

    def print_info(self):
        print(f"{self.name} {self.age}")


# Định nghĩa lớp Student kế thừa lớp Person
class Student(Person):
    def __init__(self, name, age, ID, grade):  
        super().__init__(name, age) 
        self.ID = ID
        self.grade = grade

    def print_info(self):
        print(f"{self.name} {self.age} {self.ID} {self.grade}")
```

```python
p1 = Person("Alice", 20)
p1.print_info()
```

```python
p1 = Person("Trongdepzai", 20)
p1.print_info()
```

```python
s1 = Student("Bob", 21, "S123", "A")
s1.print_info()
```
